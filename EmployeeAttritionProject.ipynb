Method to print distinct values of any specified column

```python
def unique(df, column_name):
    # Check if the column exists in the DataFrame
    if column_name not in df.columns:
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return
    
    # Extract unique values of the specified column
    unique_values = df[column_name].unique()
    return unique_values
```

#### Deleting useless columns that have same value throughout dataset

```python
def drop_column(df, column_name):
    if column_name not in df.columns:
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return df
    
    return df.drop(columns=[column_name], axis = 1, inplace = True)
```

ALso, where there is trend of outliers showing through histogram that how these outliers are not really different from actual data and they makes sense.

```python
def create_boxplot(df, column_name):
    # Set style
    sns.set_style("whitegrid")
    
    # Create boxplot
    ax = sns.boxplot(x=df[column_name], color='turquoise', width=0.5)
    
    # Add title
    ax.set_title("Boxplot of " + column_name)
    
    # Set x-label
    ax.set_xlabel(column_name)
    
    # Set y-label
    ax.set_ylabel("Values")
    
    # Calculate and print median and IQR
    median = df[column_name].median()
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    print("Median:", median)
    print("IQR:", IQR)
    
    # Show plot
    plt.show()
```

```python
def round_to_integer(df, column):
    df[column] = df[column].round()
    
```

```python
def create_numerical_histogram(df, column_name, b):
    # Set Seaborn style
    sns.set_style("whitegrid")
    
    # Create histogram
    sns.histplot(df[column_name], bins=b, color='purple', edgecolor='black')
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    plt.gca().xaxis.set_label_coords(0.5, -0.1)
    # Add labels and title
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column_name}')
    
    # Show plot
    plt.show()
```

##### Analyzing skewness and kurtosis for some major numerical columns

```python
def calculate_skewness_and_kurtosis(df, column_name):

    sns.distplot(df[column_name])
    # Calculate skewness and kurtosis
    skewness = df[column_name].skew()
    kurt = df[column_name].kurt()
    
    print("Skewness:", skewness)
    print("Kurtosis:", kurt)
```

#### Analysis of Categorical columns through anova, t-test and chi-square.

```python
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def perform_anova_tukey(dataframe, dependent_variable, independent_variable):
    
    # Replace category names in the DataFrame
    dataframe.columns = [col.replace(' ', '_').replace('%', '_percent') for col in dataframe.columns]

    # Fit OLS model
    formula = f'{dependent_variable} ~ C(Q("{independent_variable}"))'
    model = ols(formula, data=dataframe).fit()
    
    # Perform ANOVA
    anova_table = model.summary().tables[0]
    print("ANOVA Table:")
    print(anova_table)
    
    # Check ANOVA p-value
    p_value = model.f_pvalue
    
    # Perform Tukey's test if ANOVA p-value is less than 0.05
    print ("\np value: ", p_value, '\n')
    if p_value < 0.05:
        print(f"{dependent_variable} - {independent_variable}: We reject null hypothesis because p-value < 0.05 which implies that the mean values across each group are not equal.")

        tukey_results = pairwise_tukeyhsd(dataframe[dependent_variable], dataframe[independent_variable])
        print("\nTukey's Test Results:")
        print(tukey_results)
    else:
        print(f"{dependent_variable} - {independent_variable}: We fail to reject null hypothesis because p-value > 0.05 which implies that the mean values across each group are equal.")
```

```python
from scipy.stats import ttest_ind

def perform_t_test(dataframe, group1, group2, dependent_variable, independent_variable):
    group1_data = dataframe[dataframe[independent_variable] == group1][dependent_variable]
    group2_data = dataframe[dataframe[independent_variable] == group2][dependent_variable]

    # Perform t-test
    t_statistic, p_value = ttest_ind(group1_data, group2_data)

    print(f"{independent_variable} - {group1} vs {group2} T-Test:")
    print(f"T-Statistic: {t_statistic}")
    print(f"P-Value: {p_value}")

    if p_value < 0.05:
        print(f"{dependent_variable} - {independent_variable}: We reject null hypothesis because p-value < 0.05 which implies that the mean values across each group are not equal.")
    else:
        print(f"{dependent_variable} - {independent_variable}: We fail to reject null hypothesis because p-value > 0.05 which implies that the mean values across each group are equal.")
```

```python
from scipy.stats import chi2_contingency

def chi_square_test(df, column1, column2):
    # Count of occurrences of all combinations of the two columns
    data_crosstab = pd.crosstab(df[column1], df[column2], margins=False)
    print(data_crosstab)

    # Perform chi-square test
    stat, p, dof, expected = chi2_contingency(data_crosstab)
    print('Degrees of Freedom (dof) = %d' % dof)
    print('Expected frequencies:')
    print(expected)

    # Interpret p-value
    alpha = 0.05
    print('Significance level (alpha) = %.3f' % alpha)
    print('p-value = %.3f' % p)
    if p <= alpha:
        print(f"{column1} - {column2}: Dependent (reject H0): There is evidence that the two categorical variables are associated as p < 0.05.")
    else:
        print(f"{column1} - {column2}: Independent (fail to reject H0): There is no evidence that the two categorical variables are associated as p > 0.05.")
```
